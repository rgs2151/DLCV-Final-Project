```python
class Writer:
    temp_path = ''

    def write_to_csv(self, result: ndarray, out_csv: str):
        out_csv = Path(out_csv)

        with out_csv.open('a+') as csv:
            csv_writer = writer(csv)
            csv_writer.writerow(result)

    def write_as_video(self, path, image):
        if self.temp_path != path: 
            self.temp_path = path
            self.writer = VideoWriter(path, VideoWriter_fourcc('M','J','P','G'), 10, (640, 480))

        self.writer.write(image)
```

```python
source_directory = path_to_source_folder
w = Writer()
y = []
x = []
f = []
e = []
all_source_ds =[]
x_valid =[]
y_valid = []
f_valid =[]
base_dir = "/content/new_20class_csv"
```

```python
classes = sorted(list(x.name for x in source_directory.glob('*')))

if any('.ipynb_checkpoints' in s for s in classes):
    classes.remove('.ipynb_checkpoints')
print(classes,"\n",len(classes))
```

```python
all_source_ds =[]
for category in classes:
    cat_dir = list((source_directory/category).glob("*"))
    all_source_ds.append(cat_dir)
print(len(all_source_ds),len(all_source_ds[0]))
```

```python
all_source_ds_flat = [item for sublist in all_source_ds for item in sublist]
print(len(all_source_ds_flat))
```

```python
pbar = tqdm(total=len(all_source_ds_flat), bar_format ='{l_bar}{bar}|  {n_fmt}/{total_fmt}  [ Elapsed Time : {elapsed} | ETA : {remaining} ]', leave=True, position=0)
```

```python
label =[]
inner_dir = []
for i in range(len(all_source_ds)):
    inner_dir = all_source_ds[i]
    class_name = classes[i]
    out_directory = Path(base_dir + "/" + class_name)
    if not out_directory.exists():
        out_directory.mkdir()
    for csv in inner_dir:
        d = read_csv(str(csv), header=None ).values
        if d.shape == (50,1629):
                new_csv_content = []
                csv_content =[]
                new_new_csv_content =[]
                csv_content.append(d)
                for j in range(len(csv_content[0])):
                    tempo = (csv_content[0][j][-225:])
                    new_csv_content.append(tempo[0])
                    w.write_to_csv(tempo, out_directory / f'{split(split(csv)[-1])[1][:-4]}.csv')     
        else:
                f_valid.append(d)
                print(csv,"  ",d.shape)
        pbar.update(1)
```

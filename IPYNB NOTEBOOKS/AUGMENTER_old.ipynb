```python
from pathlib import Path
from vidaug import augmentors as va
import skvideo.io
import cv2
import numpy as np
from IPython.display import HTML
from base64 import b64encode
from google.colab.patches import cv2_imshow
import random
random.seed(1)
```

```python
total_class_size = 1000

class_paths = list(Path("/content/drive/MyDrive/DATASET/ORIGINAL").glob("*"))
aug_ds = Path("/content/aug2")
```

```python
for i in class_paths:
    class_name = i.name
    videos_of_class = list((i/"video").glob("*"))

    num_to_aug = total_class_size - len(videos_of_class)

    counter_vid = num_to_aug
    for _ in range(num_to_aug):
        cap = cv2.VideoCapture(str(random.choice(videos_of_class)))
        video_frames = []

        while True:
            status, frame = cap.read()
            if not status: break
            video_frames.append(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

        video_shape = video_frames[0].shape

        seq = va.Sequential([va.RandomTranslate(int(video_shape[1]/4),50), va.RandomShear(0.07,0.07)])

        video_aug = seq(video_frames)
        (aug_ds/class_name).mkdir(exist_ok = True)
        aug_vid_name = f"{aug_ds}/{class_name}/{str(random.randint(1000000,9999999))}.avi"
        skvideo.io.vwrite(aug_vid_name, video_aug)
        print("class:",class_name,"||","vidno:",counter_vid,"/",num_to_aug,"||","filename:",aug_vid_name)
        counter_vid -=1
```
